# 2.주피터 노트북 실행법 
- shift+Enter : 포커스된 셀이 실행되고 커서가 다음셀로 이동
- alt+Enter:포커스가된 셀이 실행되고 다음셀에 생성한 후 다음셀로 이동
- ctrl+Enter:포커스된 셀이 실행되고 포커스 이동이 없음 
- Enter:다시 편집 상태로

```python
# 주석(실행 안됌)
# 셀 하나하나에 파이썬 실행 소스 입력
```

```python
# print() 함수를 이용해서 출력
print('hello, world')
print(1)
print("Hello\n\n홍길동")
```
